# create a correlation heatmap for conclusion

```python
columns = ['course_count', 'pop_over_65', 'total_pop', 'pct_over_65', 
                                      'median_income', 'land_area_sqmi']
```

```python
corr = county_metrics[columns].corr()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
